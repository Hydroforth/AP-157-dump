1. Create an $N \times N$ array with a circle inside (everything outside the "radius" of the circle is set to 0). The value of the array can be based on distance from the center.
2. Plot this array with color bars.

```python
import numpy as np
import matplotlib.pyplot as plt

def circRain(N):
    mypoints = []
    x = np.linspace(-1,1,N)
    y = np.linspace(-1,1,N)
    X, Y = np.meshgrid(x, y)

    r_center = np.sqrt(X**2 + Y**2)
    r_center[r_center > 1] = 0.0

    plt.matshow(r_center)
    plt.colorbar()
    
circRain(100)
```

```python
data = plt.imread("Pictures/macbeth.jpeg")
imgplot = plt.imshow(data)
plt.colorbar()                    
```

```python
#we obtain a 3 x 3 x 1 array, which we need to "flatten" out into 1D 
R,Rf = data[:,:,0],data[:,:,0].flatten()
G,Gf = data[:,:,1],data[:,:,1].flatten()
B,Bf = data[:,:,2],data[:,:,2].flatten()
```

```python
fig, ax = plt.subplots(2,2)
ax[0][0].imshow(data)
ax[0][1].imshow(R, cmap = "Greys_r")
ax[0][1].set_title("Red channel")
ax[1][0].imshow(G, cmap = "Greys_r")
ax[1][0].set_title("Green channel")
ax[1][1].imshow(B, cmap = "Greys_r")
ax[1][1].set_title("Blue channel")

ax[0][0].axis('off')
ax[0][1].axis('off')
ax[1][0].axis('off')
ax[1][1].axis('off')
```

```python
fig, ax = plt.subplots(figsize = (10,7))

#creates an array which function as the bins for our histogram 
binning = np.arange(0,255)

#each value from the arrays R, G, and B are binned into the preset bins defined above
ax.hist(Rf, bins=binning, histtype="step", color="red")
ax.hist(Gf, bins=binning, histtype="step", color="green")
ax.hist(Bf, bins=binning, histtype="step", color="blue")
#we set axis limits for x
ax.set_xlim(0, 255)
ax.set_xlabel("Pixel Intensity (8-bit)")
ax.set_ylabel("Pixel Count")
```

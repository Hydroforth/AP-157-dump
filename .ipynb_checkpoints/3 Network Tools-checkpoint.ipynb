#### Simple Graph

We refer to a set of interconnected nodes as a Graph. A basic graph is defined by 
\begin{align}
G = (V,E)
\end{align}
, with $G$ being the set of nodes and $E$ being the set of edges (connections between node pairs).

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

G = nx.Graph()
E = [(1,2), (2,3), (1,3), (3,4)]

G.add_edges_from(E)

print("Nodes:", G.nodes)
nx.draw(G, with_labels = True)
```

#### Directed Graphs

We can also define a graph with directions to show relationships between different nodes. A simple example would be the idea of whether or not you consider someone a friend - this does not necessitate reciprocity!

```python
G_directed = nx.DiGraph()
V_directed = ['Alice', 'Bob', 'Cate']
E_directed = [('Alice', 'Bob'), ('Bob', 'Alice'), ('Cate', 'Bob'), ('Alice', 'Cate')] 

G_directed.add_nodes_from(V_directed) #we can also use G.add_node(node)
G_directed.add_edges_from(E_directed) #we can also use G.add_edge((node,node))
nx.draw(G_directed, node_color = 'C0', node_size = 2000, with_labels = True) #beautifying the graph
```

```python
nx.draw_circular(G_directed, node_color = 'C0', node_size = 2000, with_labels = True)
```

#### Erdos-Renyi (ER) Graph

Given a set of $n$ nodes, edges are drawn between pairs of nodes with some probability $p$.

```python
#enumerate takes a list and returns a list of tuples with format (index, list[index])
#yield allows use to iterate over a sequence without storing their values

#we first take a list of all possible edges given a set of nodes V
def gen_all_pairs(V):
    for i, node1 in enumerate(V):
        for j, node2 in enumerate(V):
            if i < j: #prevents double counting
                yield node1, node2
def check(p):
    return np.random.random() < p # p probability for the edge to be part of the graph

def ER_graph(n, p):
    G = nx.Graph()
    V = range(n)
    
    for pair in gen_all_pairs(V):
        if check(p):
            G.add_edge(pair[0],pair[1])
    
    return G

G_ER = ER_graph(6,0.5)
nx.draw(G_ER, with_labels = True)
```

```python
#nx.draw_circular(G_ER, node_size = 2000, with_labels=True)
pos=nx.circular_layout(G_ER)
nx.draw(G_ER, pos, node_size = 2000, with_labels = True)
```

#### Adjacency Matrix
Graphs can be represented as an adjacency matrix by placing one node per row and one node per column. The entries represent the edges and can be a simple binary ($1$ signifying an edge between two nodes and $0$ signifying the lack thereof), or it can also be used to model a weighted graph (e.g. take values from $0$ to $5$). 

```python
karate_adj = nx.adjacency_matrix(G_karate)
print(type(karate_adj))
#print("Karate club adjacency matrix: \n" + str(karate_adj))
```

```python
karate_adj.todense()
```

```python
ER_adj = nx.adjacency_matrix(G_ER)
ER_adj.todense()
```

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

G = nx.DiGraph()

steve = [("Steve", "Ember", 4), ("Steve", "Gian", 3), ("Steve", "William", 2)]
sam = [("Sam", "CJ", 2), ("Sam", "Allen", 2), ("Sam", "Spencer", 5)]
nicole = [("Nicole","Vince",5), ("Nicole","Renz",5), ("Nicole","Marco",5)]
marco = [("Marco", "Mikael", 2), ("Marco", "Nicole", 5), ("Marco", "Allen", 4)]
mikael = [("Mikael", "Steve", 3), ("Mikael", "Vince", 3), ("Mikael", "Nicole", 3)]
ember = [("Ember","Steve",4), ("Ember","William",4), ("Ember","Daniel",3)]
allen = [("Allen", "Vince", 5), ("Allen", "Marco", 5), ("Allen", "Renz", 3)]
daniel = [("Daniel", "Marco", 4), ("Daniel", "William", 3), ("Daniel", "Nicole", 3)]
luis = [("Luis","Val",5), ("Luis","Nicole",1), ("Luis","Marco",1)]
renz = [("Renz","Nicole",1), ("Renz","Spencer",5), ("Renz","Marco",5)]
vince = [("Vince","Marco",5), ("Vince","Nicole",5), ("Vince","Allen",5)]
cj = [("CJ","Sam",2),("CJ","Marco",2),("CJ","Ember",2)]
spencer = [("Spencer","Sam",5),("Spencer","Renz",5),("Spencer","Marco",5)]
val = [("Val","Marco",1), ("Val","Nicole",1), ("Val","Luis",5)]
william = [("William","Daniel",3),("William","Ember",4),("William","Miguel",1)]

E = steve + sam + nicole + marco + mikael + ember + allen + daniel + luis + renz + vince + cj + spencer + val + william
G.add_weighted_edges_from(E)
```

```python
pos = nx.kamada_kawai_layout(G)
#pos = nx.nx_agraph.graphviz_layout(G, prog="neato")

edges,weights = zip(*nx.get_edge_attributes(G,'weight').items())

fig, ax = plt.subplots(figsize = (20,12))
cmap = plt.cm.Blues
nx.draw(G, pos, with_labels = True, node_size = 1500, arrowsize = 20, node_color = "yellow", edgelist=edges, edge_color = weights, width = 2, edge_cmap = cmap, connectionstyle="arc3,rad=0.2")
fig.set_facecolor('#e8587c')
vmin = min(weights)
vmax = max(weights)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin = vmin, vmax=vmax))
plt.colorbar(sm)
plt.show()
```

# Morphological Operations

```python
structure_el = disk(1) #from skimage
structure_el
```

```python
dilated = dilation(img_gray_bool, structure_el)
plt.imshow(dilated, interpolation = 'none')
plt.colorbar()
```

```python
dilated
```

# Blob Detection

```python
blobs = blob_log(dilated, max_sigma=30, min_sigma = 23, num_sigma=2, threshold = .3, overlap = 0.1)
```

```python
fig, ax = plt.subplots(figsize = (6,6))
ax.imshow(dilated, interpolation = 'none')
for blob in blobs:
    y, x, area = blob
    ax.add_patch(plt.Circle((x,y), area*np.sqrt(2), color = 'r', fill = False))
    
```
